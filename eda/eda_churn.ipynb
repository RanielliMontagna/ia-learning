## Bivariate Analysis

```python
# Hipothesis: Clients with monthlty contract are more likely to churn
# Create a contingency table with Contract and Churn variables - Quantity
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
# Hipothesis: Clients with monthlty contract are more likely to churn
# Create a contingency table with Contract and Churn variables - Percentage
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize='index', margins=True, margins_name='Total')
```

Considering all clients, the monthly contract has a higher churn rate, but if we consider only the clients who have churned, the churn rate is higher in the annual contract.

```python
# Correlation between two variables (qualitative)

# Execute a test called Chi-Square or Qui-Square Pearson
# In a hypothesis test, two hypotheses are formulated:
# H0 (Null hypothesis): There is no association between the variables
# H1 (Alternative hypothesis): There is association between the variables

# The test is used to confirm or reject the null hypothesis
# When the probability of observing H0 is less than 5% (p-value <0.05), we reject H0
# Rejecting H0 means that there is association between the variables (H1)

```

```python
# Generate a DF Crosstab (withou margins)
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Calculate scores and p-value
from scipy.stats import chi2_contingency
```

```python
# Tuple = Imutable List
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Show numbers witsh decimal place withou scientific notation
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Create a DataFrame with scores and p-values
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
# Analyze Scores and p-values
df_chi_scores_churn_contract
```

- Conclusion: There is association between Churn and Contract variables (p-value < 0.05)

```python
# Evaluate the correlation between a qualitative variable and a quantitative variable
# Use Chi-Square test, but first, we need to transform the quantitative variable into a qualitative variable

# Hipothesis: Clients with less than 6 months of contract are more likely to churn

# Create a new column with tenure transformed in qualitative variable
# Column = 'LessThan6Months'

import numpy as np
```

```python
# Create new column
df_churn['LessThan6Months'] = np.where(df_churn['tenure'] < 6, 'Yes', 'No')
```

```python
# Check dataframe
df_churn.head(5)
```

```python
# Generate a DF Crosstab (withou margins)
df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.LessThan6Months)
```

```python
# Calculate scores and p-value
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
# Create a DataFrame with scores and p-values
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p-value': pvalues_churn_tenure})
```

```python
# Analyze Scores and p-values
df_chi_scores_churn_tenure
```

As the p-value is less than 0.05, we reject H0, that is, the variables are not independent. And it is worth mentioning that based on the Qui2 resulting from the 2 analyzes, the correlation between Churn and Contract Time <6 months is less than the correlation between Churn and Contract Type

```python
# Create a dataframe with only clients with month-to-month contract
df_churn_monthly = df_churn[df_churn.Contract == 'Month-to-month']
```

```python
df_churn_monthly.info()
```

```python
# Histogram - Check visual distribution of TotalCharges
df_churn_monthly.TotalCharges.plot.hist()
```

```python
# Turkey Method - IQR (Anomaly Distribution)
# IQR - Interquartile Range
# IQR = Q3 - Q1
q1_TotalCharges_monthly = df_churn_monthly.TotalCharges.quantile(0.25)
q3_TotalCharges_monthly = df_churn_monthly.TotalCharges.quantile(0.75)

iqr_TotalCharges_monthly = q3_TotalCharges_monthly - q1_TotalCharges_monthly

iqr_TotalCharges_monthly
```

```python
# Calculate the lower and upper limits
lower_limit_TotalCharges_monthly = q1_TotalCharges_monthly - 1.5 * iqr_TotalCharges_monthly
upper_limit_TotalCharges_monthly = q3_TotalCharges_monthly + 1.5 * iqr_TotalCharges_monthly
lower_limit_TotalCharges_monthly, upper_limit_TotalCharges_monthly
```

```python
df_churn_monthly[
    (df_churn_monthly.TotalCharges < lower_limit_TotalCharges_monthly) |
    (df_churn_monthly.TotalCharges > upper_limit_TotalCharges_monthly)
]
```

```python
# Z-Score Method - Use for normal distribution
# Indicates how many standard deviations an observation is from the mean
# Z-Score = (observation - mean) / standard deviation
from scipy.stats import zscore

z = np.abs(zscore(df_churn_monthly.TotalCharges))
```

```python
z
```

```python
df_churn_monthly[z > 3.0]
```
